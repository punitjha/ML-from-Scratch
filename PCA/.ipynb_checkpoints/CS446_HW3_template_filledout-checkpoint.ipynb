### Image compression

The SVD allows us to compress an image by throwing away the least important information.  The greater the singular values -> the greater the variance -> most information from the corresponding singular vector.

Perform SVD on the matrix and get rid of the small singular values to compress the image. The loss of information is negligible as the difference is very difficult to spot.

Each singular value tells us how much of the variance of a matrix (e.g. image) is captured in each component.  For example, the variance captured by the first component is $$\frac{\sigma_1}{\sum_{i=1}^n \sigma_i}$$ where $\sigma_i$ is the $i^{th}$ singular value.

You need to finish the following functions to do SVD and then reconstruct the image by components.

```python
def svd(X):
    """
    Perform SVD. You can use numpy's svd method.

    We assume that each column of the image (input matrix X) is a feature.
    
    Args: 
        X: N * D array corresponding to an image
    Return:
        U: N*N
        S: min(N, D)*1
        V: D*D
    """
    U, S, V = np.linalg.svd(X, compute_uv=True)
    return U, S, V
    

def rebuildsvd(U, S, V, k):
    """
    Rebuild SVD by k componments.
    Args: 
        U: N*N
        S: min(N, D)*1
        V: D*D 
        k: int corresponding to number of components
    Return:
        Xrebuild: N*D array of reconstructed image
    """
    rebuilt = np.zeros((U.shape[0], V.shape[1]))
    for i in range(0, k):
        new_component = S[i] * np.dot(U[:, i, np.newaxis], V[i, :, np.newaxis,].T)
        rebuilt += new_component
    return rebuilt

def compression_ratio(X, k):
    """
    Compute compression of an image: 
        (number of stored values in original)/(number of stored values in compressed)
    Args: 
        X: N * D array corresponding to an image
        k: int corresponding to number of components
    Return:
        compression_ratio: float of proportion of storage used by compressed image
    """
    numX = np.prod(X.shape)
    numCX = k * X.shape[0] + k + k * X.shape[1]
    return float(numX) / float(numCX) 

def recovered_variance_proportion(S, k):
    """
    Compute the proportion of the variance in the original matrix recovered by a rank-k approximation
    
    Args:
       S: min(N, D)*1 of singular values for the image
       k: int, rank of approximation
    Return:
       recovered_var: int corresponding to proportion of recovered variance
    """
    usedS = S[0:k]
    return np.sum(usedS) / np.sum(S)
```

**Intrinsic Dimensionality [1pt]**  
Assume a dataset is composed of N datapoints, each of which has D features with D < N.  The *dimension* of our data would be $D$. It is possible, however, that many of these dimensions contain redundant information.  The *intrinsic dimensionality* is the number of dimensions we need to reconstruct our data with high fidelity.  For our purposes, we will define the intrinsic dimension as the number of principal components needed to reconstruct 99% of the variation within our data.  

We define a set of features as linearly independent if we cannot construct one out of a linear combination of the others. The number of linearly independent features is the number of nonzero principal components (where we define 0 is anything less than $10^{-11}$ due to floating point error). Zero principal components mean that we can not find any weights to linearly combine features in order to create an independent feature. Thus, our algorithm will assign 0 to these weights.

```python
def intrinsic_dimension(S, recovered_variance=.99):
    """
    Find the number of principal components necessary to recover given proportion of variance
    
    Args: 
        S: 1-d array corresponding to the singular values of a dataset
        
        recovered_varaiance: float in [0,1].  Minimum amount of variance 
            to recover from given principal components
    Return:
        dim: int, the number of principal components necessary to recover 
            the given proportion of the variance
    """
    summ = 0
    denom = np.sum(S)
    for i in range(S.size):
        if recovered_variance_proportion(S, i) >= recovered_variance:#summ >= recovered_variance:
            break
    return i+1
    
def num_linearly_ind_features(S, eps=1e-11):
    """
    Find the number of linearly independent featues in dataset
    
    Args: 
        S: 1-d array corresponding to the singular values of a dataset
    Return:
        dim: int, the number of linearly independent dimensions in our data
    """
    num = 0
    for i in range(S.size):
        if S[i] > eps:
            num += 1
    return num
```

**Feature Scaling and Centering [2pts]**

Principal component analysis is not agnostic to the scale and mean of your features.  Measuring a feature with different units can change your principal components.

For this problem, implement ``randomly_perturb_data`` by randomly choosing one of your columns and multiplying it by 1000. Also, implement ``center_data`` to give each column zero mean.
```python
def randomly_perturb_data(data, multiplier=1000):
    """
    Multiply a random column in data by multiplier
    
    Inputs:
        data: N*D numpy array of features
        multiplier: multiplier by which to perturb a random column in data
        
    Returns:
        perturbed_data: Data with random column multiplied by multiplier
    """
    datam = np.copy(data)
    col = np.random.randint(0, high=datam.shape[1])
    print("Perturbed column: " + str(col))
    
    datam[:, col] = np.dot(datam[:,col],multiplier)
    return datam

def center_data(data):
    """
    Make the features of the data have a mean of zero.
    
    Inputs:
        data: N*D numpy array of features
        
    Returns:
        centered_data: Data centered about the origin
    """
    datam = np.copy(data)
    datam = datam - np.reshape(np.mean(datam, axis=0), (1, datam.shape[1]))
    return datam

def scree_plot(S, color, n_components=10):
    """
    Plot proportion of variance contained in each individual component
    """
    r = np.arange(0,n_components)
    plt.plot(r + 1, S[:n_components]/np.sum(S), '-', color=color)
    
def plot_component_vector(V, color, num):
    x = np.arange(0, V.shape[1]* 3)
    x = x[num:V.shape[1]*3:3]
    plt.bar(x, np.abs(V[0,:]), color=color)
        
def plot_pca_analysis(data):
    perturbed = randomly_perturb_data(data)
    centered = center_data(data)
    U,S,V = svd(data)
    Up,Sp,Vp = svd(perturbed)
    Uc,Sc,Vc = svd(centered)
    scree_plot(S, 'b')
    scree_plot(Sp, 'r')
    scree_plot(Sc, 'g')
    plt.figure()
    plot_component_vector(V, 'b', 0)
    plot_component_vector(Vp, 'r', 1)
    plot_component_vector(Vc, 'g', 2)
    print("Original number of linearly indepenent features: " + str(num_linearly_ind_features(S)))
    print("Perturbed number of linearly indepenent features: " + str(num_linearly_ind_features(Sp)))
    print("Centered number of linearly indepenent features: " + str(num_linearly_ind_features(Sc)))
    print("Original intrinsic dimensionality: " + str(intrinsic_dimension(S)))
    print("Perturbed intrinsic dimensionality: " + str(intrinsic_dimension(Sp)))
    print("Centered intrinsic dimensionality: " + str(intrinsic_dimension(Sc)))
    plt.figure()
```
